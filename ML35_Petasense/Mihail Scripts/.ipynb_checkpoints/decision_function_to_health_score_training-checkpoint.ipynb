```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from local_outlier_factor import LocalOutlierFactor
```

```python
base_dir = 'features/'

locations = os.listdir(base_dir)
for location in locations:
    location_dir = os.path.join(base_dir, location)
    
    ### Read datasets
    # Baseline
    baseline = pd.read_csv(os.path.join(location_dir, 'baseline.csv'))
    
    # Calibration sets
    warning = pd.read_csv(os.path.join(location_dir, 'warning.csv'))
    critical = pd.read_csv(os.path.join(location_dir, 'critical.csv'))
    worst = pd.read_csv(os.path.join(location_dir, 'worst.csv'))
    
    ### Fit LOF on baseline
    model = LocalOutlierFactor()
    model.fit(baseline)
    
    ### Compute LOF decision function values on the baseline and calibration sets
    baseline_scores = model.decision_function(baseline)
    warning_scores = model.decision_function(warning)
    critical_scores = model.decision_function(critical)
    worst_scores = model.decision_function(worst)
    
    ### Calibrate function that maps decision function to health score
    # Create the training set of the current location
    labeling_jitter = True
    if labeling_jitter:
        training_set = np.vstack((
            np.stack((np.random.uniform(8.95, 9.05, len(baseline_scores)), baseline_scores), axis=1),
            np.stack((np.random.uniform(6.95, 7.05, len(warning_scores)), warning_scores), axis=1),
            np.stack((np.random.uniform(3.95, 4.05, len(critical_scores)), critical_scores), axis=1),
            np.stack((np.random.uniform(0, 0.05, len(worst_scores)), worst_scores), axis=1)))
    else:
        training_set = np.vstack((
            np.stack((np.repeat(9, len(baseline_scores)), baseline_scores), axis=1),
            np.stack((np.repeat(7, len(warning_scores)), warning_scores), axis=1),
            np.stack((np.repeat(4, len(critical_scores)), critical_scores), axis=1),
            np.stack((np.repeat(0, len(worst_scores)), worst_scores), axis=1)))
    
    # Decision function to health score mapping functions
    def polynomial(x, F):
        return np.polyval(F, x)
    
    def logistic(x, L = 9, k = 0.75, x0 = 5):
        return L/(1 + np.exp(-k*(x-x0)))
    
    def logistic_polynomial(x, F):
        return logistic(polynomial(x, F))
    
    # Fit polynomial curve from decision function to labels
    # Degree of polynomial, third degree seems to work best
    polynomial_degree = 3
    F = np.polyfit(training_set[:,1], training_set[:,0], deg=polynomial_degree)
    
    ### Plot training data
    # Training plots toggle
    training_plots = True
    
    if training_plots:
        # Plot training set
        plt.figure(figsize=(9,5))
        plt.title('Location id: '+location)
        plt.ylabel('Health score')
        plt.xlabel('Decision function')
        plt.scatter(training_set[:,1], training_set[:,0])
        plt.show()
    
    ### Plot mapping functions in decision function domain
    x = np.linspace(np.min(training_set[:,1]), np.max(training_set[:,1]), 100)
    plt.figure(figsize=(9,5))
    plt.subplot(1,2,1)
    plt.title('Polynomial, degree '+str(polynomial_degree))
    plt.ylabel('Health score')
    plt.xlabel('Decision function')
    plt.plot(x, polynomial(x, F))
    plt.subplot(1,2,2)
    plt.title('Logistic polynomial')
    plt.ylabel('Health score')
    plt.xlabel('Decision function')
    plt.plot(x, logistic_polynomial(x, F))
    plt.show()
    
    ### Plot training set decision function to health score mappings
    plt.figure(figsize=(9,5))
    plt.title('Health score mappings')
    plt.plot(polynomial(training_set[:,1], F))
    plt.plot(logistic_polynomial(training_set[:,1], F))
    plt.ylabel('Health score')
    plt.legend(['Polynomial', 'Logistic polynomial'])
    plt.show()
```
